Tidying Up Period

```python
df[['period_group', 'period_specific']] = df['Period'].str.split(',', expand=True)
```

```python
period_groupby = df.groupby('period_group')['period_group'].count().sort_values(ascending=False)
period_groupby.plot(kind = 'bar', title = 'Dinosaur Period', 
                  color = cmap(np.linspace(0, 1, type_grouped.nunique() + 1)), figsize = (10,5))
plt.title('Dinosaur Period of Living')
plt.xlabel('Living Period')
plt.ylabel('# Dinosaurs')
plt.xticks(rotation=30, horizontalalignment="center")


plt.show()
```

```python
df.groupby('period_specific')['period_specific'].count().sort_values(ascending=False)
```

```python
#split exact number of living period
df[['earliest_period', 'latest_period']] = df['period_specific'].str.split(' ', expand = True)[1].str.split('-', expand = True)
```

```python
df['earliest_period'].unique()
```

```python
df.loc[df['earliest_period'].isnull()]
```

```python
df.loc[df['latest_period'].isnull()]
```

```python
#to count estimate period of living, needs to fill null value, say with 0
```

```python
df['earliest_period'] = df['earliest_period'].fillna('0')
df['latest_period'] = df['latest_period'].fillna('0')
```

```python
df['earliest_period'] = pd.to_numeric(df['earliest_period'], errors='coerce')
df['latest_period'] = pd.to_numeric(df['latest_period'], errors='coerce')
```

```python
df['estimated_living_duration'] = df.apply(lambda row: (row['earliest_period'] - row['latest_period']),
                                          axis = 1)
```

```python
df['estimated_living_duration'] = df['estimated_living_duration'].abs()
```

```python
df['Taxonomy'].unique()
```

```python
#change format of taxonomy to make it easier to read
#df['Taxonomy'] = df['Taxonomy'].replace(',',' >')
df['Taxonomy'] = df['Taxonomy'].apply(lambda x: x.replace(',',' >'))
```

```python
df['Taxonomy'].unique()
```

```python
taxo_split = df['Taxonomy'].str.split('>', expand = True)
taxo_split = taxo_split.add_prefix('taxonomy_level_')
```

```python
df = pd.concat([df, taxo_split], axis = 1, sort = False)
```
